# compare_classifiers Package Tutorial

In this tutorial, we will demonstrate how to use the `compare_classifiers` package to compare metrics for machine learning models and decide whether to use a single Classifier or stack multiple Classifiers through **Voting** or **Stacking**.  
There are four functions in this package and each serves a unique purpose. You can use confusion matrices and `f1` scores to compare model performance and eventually predict on unseen data. 

## Getting Started

Before getting to the functions, to get started, let's set up our package environment first:  

```python
import compare_classifiers
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

We will then load an example data, the popular iris dataset. 

```python
# Load the dataset
iris = load_iris()
X, y = iris.data, iris.target
```

## Plot Confusion Matrices 

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
estimators = [
    ('Random Forest', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('SVM', make_pipeline(StandardScaler(), SVC(random_state=42)))
]
```
