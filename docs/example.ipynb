# compare_classifiers Package Tutorial

In this tutorial, we will demonstrate how to use the `compare_classifiers` package to compare metrics for machine learning models and decide whether to use a single Classifier or stack multiple Classifiers through **Voting** or **Stacking**.  
There are four functions in this package and each serves a unique purpose. You can use confusion matrices and `f1` scores to compare model performance and eventually predict on unseen data.
